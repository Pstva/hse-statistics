### Вариант 8


Пусть $$x_1, x_2, ..., x_n \mathtt{\sim} Exp(\alpha)$$
Продемонстрируйте, что распределение $$\alpha x_{[n]} - \ln(n)$$ слабо сходится к распределению с функцией распределения $$e^{-e^{-x}}, x \in \mathbb{R}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import math
from IPython.display import display, Math
import scipy.stats as st
rcParams.update({'font.size': 13})
```

Для того, чтобы показать слабую сходимость, я покажу, что функции распределения двух распределений сходятся с ростом n и покажу это для разных значений параметра $\alpha$.

Сначала приведу объяснения своих подсчетов функции распределения случайной величины 

$$\xi = \alpha x_{[n]} - \ln(n)$$

$$F_{\xi}(C) = P(\xi < C) = P(\alpha x_{[n]} - \ln(n) < C) = P(x_{[n]} < \frac{\ln(n) + C}{\alpha}) = P(x_1 < \frac{\ln(n) + C}{\alpha}, x_2 < \frac{\ln(n) + C}{\alpha}, ..., x_n < \frac{\ln(n) + C}{\alpha}) = F_e^n(\frac{\ln(n) + C}{\alpha})$$

Соответственно, я считаю значение функции распределения для C из диапазона [-3.5, 3.5] и возвожу в степень вероятности, после чего получаю искомые значения для функции распределения новой случайной величины.


### Результаты

$$\alpha = 1$$

```python
plt.figure(figsize=(15, 15))

alpha=1

X_axis = np.linspace(-3.5, 3.5, 100)

for k, n in enumerate([1, 5, 10, 15, 20, 50]):

    #P(X[n] < C) = F^n(C)
    probs = [st.expon.cdf((1/alpha) * (i + math.log(n)), scale=1/alpha) for i in X_axis]
    probs = [p**n for p in probs]
     
    plt.subplot(3, 3, k+1)
    plt.xlim([-3.5, 3.5])
    plt.tight_layout()

    plt.scatter(X_axis, probs,
        label= "$a x_{[n]} - \ln(n)$", s=2, alpha=0.8, c="purple")
    
    plt.plot(X_axis, [math.exp(-math.exp(-i)) for i in X_axis],  label='$e^{-e^{-x}}$',alpha=0.8)
    plt.xlabel(f"n: {n}", fontsize=14)
    plt.ylabel('F(X)', fontsize=12)
    plt.legend()
    
plt.show()
```
